```python
import requests
import json


base_url = 'http://127.0.0.1:5000'


def print_response(response):
    print(json.dumps(response.json(), indent=2, ensure_ascii=False))







```

```python
response = requests.get(f'{base_url}/datos')
print("Todos los datos:")
print_response(response)




```

```python
response = requests.get(f'{base_url}/datos/train')
print("Datos de entrenamiento:")
print_response(response)

```

```python
try:
    response = requests.get(f'{base_url}/datos/test')
    print("Datos de prueba:")
    print_response(response)
except requests.exceptions.RequestException as e:
    print(f"HTTP Request failed: {e}")
except json.JSONDecodeError as e:
    print(f"JSON Decode Error: {e}")
    print("Response content:")
    print(response.text)
except Exception as e:
    print(f"Error: {e}")
    if response is not None:
        print("Response content:")
        print(response.text)

```

```python
n = 0  
response = requests.get(f'{base_url}/datos/{n}')
print(f"Observación {n}:")
print_response(response)

```

```python
try:
    observacion = {
        "ID_Cliente": 1, 
        "Total": 1, 
        "Ingresos": 30000, 
        "Tamaño_Hogar": 3, 
        "Edad_26-35": 0, 
        "Edad_36-45": 1, 
        "Edad_46-55": 0, 
        "Edad_56-65": 0, 
        "Edad_65+": 0, 
        "Sexo_M": 1, 
        "Estado civil_Casado": 0, 
        "Estado civil_Divorciado": 0, 
        "Estado civil_Viudo": 0, 
        "Nivel_Educativo_Secundaria": 0, 
        "Nivel_Educativo_Universitario": 1, 
        "Nivel_Educativo_Postgrado": 0
    } 
    
    
    columnas_esperadas = requests.get(f'{base_url}/columnas_modelo').json()
    observacion_alineada = {col: observacion.get(col, 0) for col in columnas_esperadas}
    
    response = requests.post(f'{base_url}/predict', json=observacion_alineada)
    print("Predicción:")
    print_response(response)
except requests.exceptions.RequestException as e:
    print(f"HTTP Request failed: {e}")
except json.JSONDecodeError as e:
    print(f"JSON Decode Error: {e}")
    print("Response content:")
    print(response.text)
except Exception as e:
    print(f"Error: {e}")
    if response is not None:
        print("Response content:")
        print(response.text)

```
